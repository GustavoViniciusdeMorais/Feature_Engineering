```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create two simple continuous variables
x = np.random.normal(size=100)
y = np.random.normal(size=100)
```

```python
# Creating a grid figure with matplotlib
fig, my_grid = plt.subplots(nrows=2, ncols=3, figsize=(18,8)) # fig with 2 row and 2 columns

# scatter plot
sns.scatterplot(x, y, color="g", ax=my_grid[0,0])

#box plot
sns.boxplot(x, color="g", ax=my_grid[0,1])

#histogram plot
sns.distplot(x, kde=False, rug=True, color="r", ax=my_grid[1,0])

# density plot
sns.distplot(x, hist=False, rug=True, ax=my_grid[1,1])

# distribution plot
sns.distplot(x, ax=my_grid[0,2])

# strip plot
sns.stripplot(x, y, size = 10, ax=my_grid[1,2])
```
